```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
arr = np. zeros ((512 , 512 , 3))
arr [:, :] = np. array ([0.5 , 0, 1])
arr [200: , :] = np. array ([1 , 1, 1])
arr [200:300 , :] = np. array ([1,0,1])
(x, y) = np. meshgrid (np.arange (512) , np.arange (512))
x_0 = 60
y_0 = 100
c_x = (x - x_0 )**2
c_y = (y - y_0 )**2
inside = np.sqrt (c_x + c_y ) < 30
arr[inside , :] = np. array ([1 , 1, 0])
plt. imshow (arr)
plt. show ()
```
