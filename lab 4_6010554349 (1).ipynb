```
import cv2
import numpy as np
from matplotlib import pyplot as plt

#img = cv2.imread('images/download.jpg')
img = cv2.imread('FireHose1.jpg',0)
img = cv2.medianBlur(img,5)
cimg = cv2.imread('FireHose1.jpg',1)
cimg = cv2.cvtColor(cimg, cv2.COLOR_BGR2RGB)
plt.imshow(cimg)
plt.show()

circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,2000,param1=50,param2=50,minRadius=450,maxRadius=500)

circles = np.uint16(np.around(circles))

for i in circles[0,:]:
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),5)
print("The fire hose wheel is located at x = ", circles[0][0][0], ",", "y =", circles[0][0][1], " and radius =", circles[0][0][2])

plt.imshow(cimg)
```

The image have less noise so we can detected the fire hose's easily.
